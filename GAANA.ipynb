# 🎵 Songs Dataset

## 🧾 Overview
This dataset contains metadata for **41,355 songs**, covering details such as song names, singers, duration, and language. It can be useful for projects related to **music recommendation systems**, **artist analysis**, or **language-based song classification**.

---

## 📂 Dataset Information

| Feature | Description | Data Type | Example |
|----------|--------------|------------|----------|
| **name** | Name of the song | `object` | `Dil - E - Nadan Tujhe` |
| **singer** | Names of the artists/singers separated by `|` | `object` | `Chitra Singh|Jagjit Singh` |
| **singer_id** | Unique identifier(s) for the artist(s), separated by `|` | `object` | `/artist/chitra-singh|/artist/jagjeet-singh-1` |
| **duration** | Duration of the song in minutes and seconds (MM:SS) | `object` | `05:00` |
| **link** | Relative URL to the song’s page | `object` | `/dil-e-nadan-tujhe-3` |
| **language** | Language of the song | `object` | `Urdu` |

---

```python
duration_in_sec = []

for i in df['duration']:
  mins = int(i.split(":")[0])
  sec = int(i.split(":")[1])
  dur = mins*60 + sec
  duration_in_sec.append(dur)
```

```python
df.head()
```

```python
df['duration_in_secs'].describe()
```

```python
for i in df['duration']:
  if len(i.split(":")) == 3:
    print(i)
```

```python
df[df['duration'] == '07:08:00']
```

```python
duration_in_sec = []

for i in df['duration']:
  if len(i.split(":")) == 2:
    mins = int(i.split(":")[0])
    sec = int(i.split(":")[1])
    dur = mins*60 + sec
    duration_in_sec.append(dur)

  elif len(i.split(":")) == 3:
    hrs = int(i.split(":")[0])
    mins = int(i.split(":")[1])
    sec = int(i.split(":")[2])
    dur = hrs*3600 + mins*60 + sec
    duration_in_sec.append(dur)
```

```python
df['duration_in_secs'] = duration_in_sec
```
